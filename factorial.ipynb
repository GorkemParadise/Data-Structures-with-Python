```python
def faktoriyel(n):
        if n == 0 or n == 1:
            return 1
        else: 
            return n * faktoriyel(n-1)
```

```python
faktoriyel(4)
```

```python
faktoriyel(6)
```

```python
def factorial2(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
factorial2(4) # 1 2 3 4'e kadar gittik ve result * gittiğimiz sayılar
```

```python
factorial2(6)
```
