```python
arr = [80, 20, 10, 30, 60, 50, 70, 40]
```

```python
def selection_sort_min(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        
        for j in range(i+1, n):
            if arr[j] < arr[min_index]:
                min_index = j
        
        if i != min_index:
            arr[i], arr[min_index] = arr[min_index], arr[i]
    
    return arr
```

```python
selection_sort_min(arr)
```
