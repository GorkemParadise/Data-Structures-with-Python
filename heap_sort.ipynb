```python
arr = [80, 20, 10, 30, 60, 50, 70, 40]
```

```python
def heapify(arr, size, i):
    largest = i
    left = 2*i + 1
    right = 2*i + 2

    if left < size and arr[left] > arr[largest]:
        largest = left
    if right < size and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, size, largest)
```

```python
def heap_sort(arr):
    size = len(arr)

    # build max heap
    for i in range(size//2 - 1, -1, -1):
        heapify(arr, size, i)

    # extract elements
    for i in range(size-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

    return arr
```

```python
heap_sort(arr)
```
