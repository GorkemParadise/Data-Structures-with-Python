```python
arr = [80, 20, 10, 30, 60, 50, 70, 40]
```

```python
def selection_sort_max(arr):
    n = len(arr)
    for i in range(n-1, -1, -1):
        max_index = i
        
        for j in range(i):
            if arr[j] > arr[max_index]:
                max_index = j
        
        if i != max_index:
            arr[i], arr[max_index] = arr[max_index], arr[i]
    
    return arr
```

```python
selection_sort_max(arr)
```
