```python
def sum(n):
    if n == 0:
        return n
    else:
        return n + sum(n-1)
```

```python
sum(4) # 4+3+2+1 = 10
```

```python
sum(5)
```

```python
def gauss(n):
    return n * (n + 1) // 2
```

```python
gauss(5)
```

```python
def gaus2(n):
    result = 0
    for i in range(1, n + 1):
        result += i
    return result
```

```python
gaus2(5)
```

```python
gaus2(6)
```
